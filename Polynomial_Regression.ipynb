**Polynomial Regression:**

Polynomial regression is a form of regression analysis that represents the relationship between a dependent (or target) variable and an independent (or feature) variable using an 𝑛 n-degree polynomial function.

This approach is particularly valuable in machine learning for capturing complex, non-linear patterns within data, which is beneficial for datasets where the relationship between variables does not align linearly.

To check the polynomial Regression practically, lets plot the relationship between Water Quality Index and Nutrient levels and see how the models fits a non-linear line to find the relation ship between those variables and make predictions.

**Step 1: Import Required Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
```

NumPy: Essential for performing numerical operations.

Matplotlib: Enables data visualization.

scikit-learn modules: Provides tools to create polynomial features and evaluate performance.

**Step 2: Generating the Dataset**

```python
# Seed for reproducibility
np.random.seed(42)

# Generating random nutrient levels (between 0 and 20 mg/L)
nutrient_levels = np.random.uniform(0, 15, 100)

# Calculating water quality index based on the cubic relationship with some noise
water_quality_index = 0.05 * nutrient_levels**3 - 0.5 * nutrient_levels**2 + nutrient_levels + 30


# Adding random noise to simulate real-world data
noise = np.random.normal(0, 5, water_quality_index.shape)
water_quality_index += noise

# Clipping values to keep WQI within a reasonable range (0 to 100)
water_quality_index = np.clip(water_quality_index, 0, 100)
```

**Step 3: Train the Polynomial Regression Model**

```python
# Fitting a cubic polynomial
coefficients = np.polyfit(nutrient_levels, water_quality_index, 3)
polynomial = np.poly1d(coefficients)
```

**Step 4: Make Predictions**

Let’s make predictions based on the model and see how well it fits the data.

```python
# Generating x values for the polynomial curve
x_values = np.linspace(0, 20, 100)
y_values = polynomial(x_values)
```

**Step 5: Evaluate the Model**

To understand the model’s accuracy, calculate metrics like Mean Squared Error (MSE) and R-squared (R²).

```python
# Performance metrics
predicted_values = polynomial(nutrient_levels)
# R-squared
r_squared = r2_score(water_quality_index, predicted_values)
# Mean Squared Error
mse = mean_squared_error(water_quality_index, predicted_values)
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared: {r_squared:.4f}")
```

**Step 6: Visualize the Results**

Let’s plot the actual and predicted Nutrient levels against the Water Quality Index.

```python
# Plotting the data points
plt.scatter(nutrient_levels, water_quality_index, color='blue', label='Data Points')

# Plotting the polynomial curve
plt.plot(x_values, y_values, color='red', label='Cubic Fit')

# Labels and title
plt.xlabel('Nutrient Levels (mg/L)')
plt.ylabel('Water Quality Index (WQI)')
plt.title('Water Quality vs. Nutrient Levels')
plt.legend()
plt.grid(True)
plt.show()
```
